```python
import networkx as nx
import numpy as np

bytes = []

with open('inputs/day18.txt') as input:
    for line in input.readlines():
        bytes.append(tuple(int(x) for x in line.strip().split(',')))

xmax, ymax = 71, 71

def in_grid(x, y):
    return 0 <= x < xmax and 0 <= y < ymax

def corrupted(x, y, ns):
    return (x, y) in bytes[:ns]

def print_grid(ns):
    grid = np.full((xmax, ymax), '.', dtype='U')
    for x, y in bytes[:ns]:
        grid[y][x] = '#'
    for y in range(ymax):
        for x in range(xmax):
            print(grid[y, x], end='')
        print()

ns = 1024

graph = nx.Graph()
for x in range(xmax):
    for y in range(ymax):
        if corrupted(x, y, ns): continue
        for dx, dy in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
            xp, yp = x + dx, y + dy
            if corrupted(xp, yp, ns): continue
            if in_grid(xp, yp):
                graph.add_edge((x, y), (xp, yp))
print_grid(ns)
nx.shortest_path_length(graph, (0, 0), (xmax - 1, ymax - 1))
```

```python
for i in range(1024, len(bytes)):
    graph.remove_node(bytes[i])
    try:
        nx.shortest_path_length(graph, (0, 0), (xmax - 1, ymax - 1))
    except nx.NetworkXNoPath:
        print(bytes[i])
        break
```
